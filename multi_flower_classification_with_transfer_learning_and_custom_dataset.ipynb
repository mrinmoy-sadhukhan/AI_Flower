```
from detecto import core, utils, visualize

dataset = core.Dataset('flower/')
model = core.Model(['rose','sunflower','daisy','tulip','dandelion'])

model.fit(dataset)
```

```
# Specify the path to your image
from detecto import core, utils, visualize
image = utils.read_image('/content/drive/MyDrive/kaggle/test/beautiful-flower-yellow-sunflower-red-600w.jpg')
predictions = model.predict(image)
# predictions format: (labels, boxes, scores)
labels, boxes, scores = predictions
# ['alien', 'bat', 'bat']
print(labels) 
print(boxes)
# tensor([0.9952, 0.9837, 0.5153])
print(scores)
#print(scores[0])
visualize.show_labeled_image(image, boxes, labels)
```
